```python
# Step 2: Import libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
# Step 3: Load a free public model
model_name = "tiiuae/falcon-7b-instruct"  # Free model from Hugging Face
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16
)

print("🤖 Your Math & Science AI Robot is ready! Ask any question below.")
```

```python
# Step 4: Make it interactive
while True:
    question = input("\nEnter a math or science question (type 'exit' to quit): ")
    if question.lower() == "exit":
        print("Goodbye! 👋")
        break

    # Tokenize the input and generate answer
    inputs = tokenizer(question, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=150)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Print the AI's answer
    print("\n🤖 Answer:\n", answer)
```
